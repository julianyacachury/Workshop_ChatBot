````python
# Estructura simple del uso de if Statements para un Chatbot
mensaje_usuario = input("Usuario: ").lower()

if "hola" in mensaje_usuario:
    print("Bot: ¬°Hola! Andres!")
elif "adi√≥s" in mensaje_usuario or "bye" in mensaje_usuario:
    print("Bot: ¬°Nos vemos maquinola!")
else:
    print("Bot: Que decis Walter? No se entiende")
````

````python
# Manejo de Multiples condiciones.
def procesar_saludo(mensaje):
    saludos = ["hola", "buenos d√≠as", "buenas tardes", "hey", "hi","chao"]
    mensaje_lower = mensaje.lower()

    for saludo in saludos:
        if saludo in mensaje_lower:
            return "¬°Hola! ¬øEn qu√© puedo ayudarte?"
    return None

def procesar_despedida(mensaje):
    despedidas = ["adi√≥s", "bye", "hasta luego", "nos vemos", "chao"]
    mensaje_lower = mensaje.lower()

    for despedida in despedidas:
        if despedida in mensaje_lower:
            return "¬°Hasta pronto! Que tengas un buen d√≠a."
    return None
````

# Logica de Estados.

Al cumplirse uno de los Statements el bot debe avanzar a otra parte de la conversaci√≥n, esto se realiza indicandole el nuevo estado en el que debe encontrarse.


````python
class EstadoChatbot:
    def __init__(self):
        self.contexto = "inicio"
        self.datos_usuario = {}

    def cambiar_contexto(self, nuevo_contexto):
        self.contexto = nuevo_contexto
        print(f"[DEBUG] Contexto cambiado a: {nuevo_contexto}")

# Ejemplo de uso
estado = EstadoChatbot()


def procesar_mensaje(mensaje, estado):
    if estado.contexto == "inicio":
        if "hola" in mensaje.lower():
            return "¬°Hola! ¬øEn qu√© puedo ayudarte?"
        if "productos" in mensaje.lower():
            estado.cambiar_contexto("consultando_productos")
            return "¬øQu√© tipo de producto buscas?"

    # Ejercicio  definir una funci√≥n que permita
    # consultar una base de productos por categor√≠a.
    elif estado.contexto == "consultando_productos":
        # L√≥gica espec√≠fica para consulta de productos
        if "limpieza" in mensaje.lower():
            return "Tenemos jabones, escobas, traperos. "
        if "tecnolog√≠a" in mensaje.lower():
            return "Tenemos Android y Iphone si lo prefiere."

        return "Lo siento, no tenemos de eso..."

````

````python
import re

# Ejemplo b√°sico: buscar patrones
mensaje = "Mi tel√©fono es 12351-456-7890"
patron_telefono = r'\d{3}-\d{3}-\d{4}'
telefono = re.search(patron_telefono, mensaje)

if telefono:
    print(f"Tel√©fono encontrado: {telefono.group()}")
````

````python
import re

class ExtractorPatrones:
    def __init__(self):
        # Patrones comunes
        self.patron_email = r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b'
        self.patron_telefono = r'(\+?1?\d{9,15})'
        self.patron_fecha = r'\b(\d{1,2})/(\d{1,2})/(\d{4})\b'
        self.patron_hora = r'\b(\d{1,2}):(\d{2})\b'
        self.patron_numero = r'\b\d+\b'
        self.patron_dinero = r'\$\d+(?:\.\d{2})?'

    def extraer_email(self, texto):
        emails = re.findall(self.patron_email, texto)
        return emails[0] if emails else None

    def extraer_telefono(self, texto):
        telefonos = re.findall(self.patron_telefono, texto)
        return telefonos[0] if telefonos else None

    def extraer_fecha(self, texto):
        fechas = re.findall(self.patron_fecha, texto)
        if fechas:
            dia, mes, a√±o = fechas[0]
            return f"{dia}/{mes}/{a√±o}"
        return None

    def extraer_numeros(self, texto):
        return re.findall(self.patron_numero, texto)

````

````python
# Ejemplo de uso
extractor = ExtractorPatrones()

mensaje_usuario = "Mi email es invalido@asd.com juanito@loslonosaurios.com y mi tel√©fono es 1234567890"
email = extractor.extraer_email(mensaje_usuario)
telefono = extractor.extraer_telefono(mensaje_usuario)

print(f"Email: {email}")
print(f"Tel√©fono: {telefono}")
````

# Manejo de Intenciones buscando RE.



````python
import re

class ManejadorIntenciones:
    def __init__(self):
        self.intenciones = {
            'saludo': [
                r'\b(hola|hello|hi|buenos d√≠as|buenas tardes|hey)\b',
                r'^(hola|hello|hi)$'
            ],
            'consulta_precio': [
                r'\b(precio|cost[ao]|cu√°nto|vale)\b.*\b(producto|art√≠culo)\b',
                r'\b(cu√°nto cuesta|precio de)\b'
            ],
            'hacer_pedido': [
                r'\b(quiero|deseo|me gustar√≠a)\b.*\b(comprar|pedir|ordenar)\b',
                r'\b(hacer un pedido|realizar pedido)\b'
            ],
            'consultar_estado': [
                r'\b(estado|d√≥nde est√°|cu√°ndo llega)\b.*\b(pedido|orden)\b',
                r'\bmis pedidos\b'
            ]
        }

    def detectar_intencion(self, mensaje):
        mensaje_lower = mensaje.lower()

        for intencion, patrones in self.intenciones.items():
            for patron in patrones:
                if re.search(patron, mensaje_lower, re.IGNORECASE):
                    return intencion

        return 'desconocida'

    def extraer_entidades(self, mensaje, intencion):
        entidades = {}

        if intencion == 'consulta_precio':
            # Buscar nombre del producto
            producto_match = re.search(r'precio de (.+?)(?:\?|$)', mensaje, re.IGNORECASE)
            if producto_match:
                entidades['producto'] = producto_match.group(1).strip()

        elif intencion == 'hacer_pedido':
            # Buscar cantidad y producto
            cantidad_match = re.search(r'(\d+)', mensaje)
            if cantidad_match:
                entidades['cantidad'] = int(cantidad_match.group(1))

        return entidades

# Ejemplo de uso
manejador = ManejadorIntenciones()

mensajes_prueba = [
    "¬øCual es el precio de el iPhone?",
    "Quiero comprar 2 laptops",
    "¬øD√≥nde est√° mi pedido?",
    "Buenos d√≠as"
]

for mensaje in mensajes_prueba:
    intencion = manejador.detectar_intencion(mensaje)
    entidades = manejador.extraer_entidades(mensaje, intencion)
    print(f"Mensaje: '{mensaje}'")
    print(f"Intenci√≥n: {intencion}")
    print(f"Entidades: {entidades}")
    print("-" * 40)
````

````python
class FlujoChatbot:

    def __init__(self):
        self.estados = {
            'inicio': self.estado_inicio,
            'menu_principal': self.estado_menu_principal,
            'navegando_productos': self.estado_navegando_productos,
            'proceso_pedido': self.estado_proceso_pedido,
            'recopilando_datos': self.estado_recopilando_datos,
            'confirmacion': self.estado_confirmacion,
            'error_estado_desconocido': self.error_estado_desconocido,
            'recomendaciones': self.recomendaciones
        }

        self.estado_actual = 'inicio'
        self.datos_sesion = {}


        # Datos de ejemplo para el chatbot
        self.productos = {
            'smartphones': [
                {'nombre': 'iPhone 15', 'precio': 999},
                {'nombre': 'Samsung Galaxy S24', 'precio': 899}
            ],
            'laptops': [
                {'nombre': 'MacBook Air M3', 'precio': 1299},
                {'nombre': 'Dell XPS 13', 'precio': 1199}
            ]
        }

    def procesar_mensaje(self, mensaje):
        """Procesa mensaje seg√∫n el estado actual"""
        print(f"\nü§ñ [ESTADO ACTUAL]: {self.estado_actual}")
        print(f"üì® [MENSAJE RECIBIDO]: {mensaje}")
        print(f"üíæ [DATOS SESI√ìN]: {self.datos_sesion}")
        print("-" * 50)

        if self.estado_actual in self.estados:
            return self.estados[self.estado_actual](mensaje)
        else:
            return self.error_estado_desconocido()

    def cambiar_estado(self, nuevo_estado):
        """Cambia el estado del chatbot"""
        if nuevo_estado in self.estados:
            print(f"üîÑ [CAMBIO DE ESTADO]: {self.estado_actual} ‚Üí {nuevo_estado}")
            self.estado_actual = nuevo_estado
        else:
            print(f"‚ùå [ERROR] Estado desconocido: {nuevo_estado}")

    def estado_inicio(self, mensaje):
        print("üìç Ejecutando: estado_inicio()")

        if any(saludo in mensaje.lower() for saludo in ['hola', 'hi', 'hello', 'buenas']):
            self.cambiar_estado('menu_principal')
            return """¬°Hola! Bienvenido a TecnoTienda üõí

¬øQu√© te gustar√≠a hacer?
1Ô∏è‚É£ Ver productos
2Ô∏è‚É£ Consultar pedido
3Ô∏è‚É£ Hablar con soporte

Responde con el n√∫mero o dime qu√© necesitas."""

        return "¬°Hola! Para comenzar, sal√∫dame con 'hola' üòä"

    def estado_menu_principal(self, mensaje):
        print("üìç Ejecutando: estado_menu_principal()")
        mensaje_lower = mensaje.lower()

        if '1' in mensaje or 'productos' in mensaje_lower:
            self.cambiar_estado('navegando_productos')
            return """üì± Nuestros productos:

üîπ Smartphones
üîπ Laptops

¬øQu√© categor√≠a te interesa?"""

        elif '2' in mensaje or 'pedido' in mensaje_lower:
            self.cambiar_estado('recopilando_datos')
            self.datos_sesion['tipo_consulta'] = 'pedido'
            return "Para consultar tu pedido, necesito tu n√∫mero de orden o email de compra."

        elif '3' in mensaje or 'soporte' in mensaje_lower:
            return "Te pondremos en contacto con un asesor."

        elif 'menu' in mensaje_lower:
            return """¬øQu√© te gustar√≠a hacer?
1Ô∏è‚É£ Ver productos
2Ô∏è‚É£ Consultar pedido
3Ô∏è‚É£ Hablar con soporte"""

        else:
            return """No entend√≠ tu selecci√≥n. Por favor elige:
1Ô∏è‚É£ Ver productos
2Ô∏è‚É£ Consultar pedido
3Ô∏è‚É£ Hablar con soporte"""

    def estado_navegando_productos(self, mensaje):
        print("üìç Ejecutando: estado_navegando_productos()")
        mensaje_lower = mensaje.lower()

        if 'smartphones' in mensaje_lower or 'tel√©fonos' in mensaje_lower or 'celulares' in mensaje_lower:
            self.datos_sesion['categoria'] = 'smartphones'
            productos_texto = self.mostrar_productos('smartphones')
            self.cambiar_estado('proceso_pedido')
            return f"""üì± Smartphones disponibles:
{productos_texto}

¬øCu√°l te interesa? Escribe el n√∫mero del producto o 'menu' para volver."""

        elif 'laptops' in mensaje_lower or 'computadoras' in mensaje_lower:
            self.datos_sesion['categoria'] = 'laptops'
            productos_texto = self.mostrar_productos('laptops')
            self.cambiar_estado('proceso_pedido')
            return f"""üíª Laptops disponibles:
{productos_texto}

¬øCu√°l te interesa? Escribe el n√∫mero del producto o 'menu' para volver."""

        elif 'menu' in mensaje_lower:
            self.cambiar_estado('menu_principal')
            return """¬øQu√© te gustar√≠a hacer?
1Ô∏è‚É£ Ver productos
2Ô∏è‚É£ Consultar pedido
3Ô∏è‚É£ Hablar con soporte"""

        else:
            return """¬øQu√© categor√≠a te interesa?
üîπ Smartphones
üîπ Laptops

O escribe 'menu' para volver al men√∫ principal."""

    def estado_proceso_pedido(self, mensaje):
        print("üìç Ejecutando: estado_proceso_pedido()")
        mensaje_lower = mensaje.lower()

        if 'menu' in mensaje_lower:
            self.cambiar_estado('menu_principal')
            return """¬øQu√© te gustar√≠a hacer?
1Ô∏è‚É£ Ver productos
2Ô∏è‚É£ Consultar pedido
3Ô∏è‚É£ Hablar con soporte"""

        # Procesar selecci√≥n de producto
        try:
            numero_producto = int(mensaje.strip()) - 1
            categoria = self.datos_sesion.get('categoria')

            if categoria and 0 <= numero_producto < len(self.productos[categoria]):
                producto = self.productos[categoria][numero_producto]
                self.datos_sesion['producto_seleccionado'] = producto
                self.cambiar_estado('recopilando_datos')
                self.datos_sesion['tipo_consulta'] = 'compra'

                return f"""‚úÖ Has seleccionado: {producto['nombre']} - ${producto['precio']}

Para procesar tu pedido necesito algunos datos:
Por favor, ingresa tu nombre completo:"""

        except (ValueError, IndexError):
            pass

        return """Por favor selecciona un producto escribiendo su n√∫mero, o escribe 'menu' para volver al men√∫ principal."""

    def estado_recopilando_datos(self, mensaje):
        print("üìç Ejecutando: estado_recopilando_datos()")

        tipo_consulta = self.datos_sesion.get('tipo_consulta')

        if tipo_consulta == 'compra':
            if 'nombre' not in self.datos_sesion:
                self.datos_sesion['nombre'] = mensaje
                return "Perfecto. Ahora necesito tu email:"

            elif 'email' not in self.datos_sesion:
                self.datos_sesion['email'] = mensaje
                return "Excelente. Por √∫ltimo, tu direcci√≥n de env√≠o:"

            elif 'direccion' not in self.datos_sesion:
                self.datos_sesion['direccion'] = mensaje
                self.cambiar_estado('confirmacion')
                producto = self.datos_sesion['producto_seleccionado']

                return f"""üìã Resumen de tu pedido:

üë§ Nombre: {self.datos_sesion['nombre']}
üìß Email: {self.datos_sesion['email']}
üìç Direcci√≥n: {self.datos_sesion['direccion']}
üõçÔ∏è Producto: {producto['nombre']}
üí∞ Precio: ${producto['precio']}

¬øConfirmas tu pedido? (S√≠/No)"""

        elif tipo_consulta == 'pedido':
            # Simular consulta de pedido
            self.cambiar_estado('menu_principal')
            return f"""üì¶ Estado de tu pedido para: {mensaje}

Pedido #12345
Estado: En tr√°nsito üöö
Fecha estimada de entrega: Ma√±ana

¬øNecesitas algo m√°s? Escribe 'menu' para volver al men√∫ principal."""

        return "No entend√≠. ¬øPuedes repetir la informaci√≥n?"

    def estado_confirmacion(self, mensaje):
        print("üìç Ejecutando: estado_confirmacion()")
        mensaje_lower = mensaje.lower()

        if 'si' in mensaje_lower or 's√≠' in mensaje_lower or 'confirmo' in mensaje_lower:
            # Limpiar datos de sesi√≥n para nuevo pedido
            producto = self.datos_sesion['producto_seleccionado']
            self.datos_sesion.clear()
            self.cambiar_estado('menu_principal')

            return f"""üéâ ¬°Pedido confirmado!

Tu pedido de {producto['nombre']} ha sido procesado exitosamente.
N√∫mero de pedido: #PED{hash(str(producto)) % 10000}

¬øHay algo m√°s en lo que pueda ayudarte?
1Ô∏è‚É£ Ver m√°s productos
2Ô∏è‚É£ Consultar pedido
3Ô∏è‚É£ Hablar con soporte"""

        elif 'no' in mensaje_lower:
            self.datos_sesion.clear()
            self.cambiar_estado('menu_principal')
            return """‚ùå Pedido cancelado.

¬øTe gustar√≠a ver otros productos o hay algo m√°s en lo que pueda ayudarte?

1Ô∏è‚É£ Ver productos
2Ô∏è‚É£ Consultar pedido
3Ô∏è‚É£ Hablar con soporte"""

        else:
            return "Por favor responde 'S√≠' para confirmar o 'No' para cancelar el pedido."

    def mostrar_productos(self, categoria):
        """Genera texto con lista de productos de una categor√≠a"""
        productos_texto = ""
        for i, producto in enumerate(self.productos[categoria], 1):
            productos_texto += f"{i}. {producto['nombre']} - ${producto['precio']}\n"
        return productos_texto

    def error_estado_desconocido(self):
        print("‚ùå [ERROR] Estado desconocido")
        return "Ha ocurrido un error. Reiniciando..."


# C√≥digo principal para ejecutar el chatbot
def ejecutar_chatbot():
    print("=" * 60)
    print("ü§ñ CHATBOT TECNOTIENDA - DEMO SIMPLIFICADA")
    print("=" * 60)
    print("Instrucciones:")
    print("- Escribe 'hola' para comenzar")
    print("- Escribe 'salir' en cualquier momento para terminar")
    print("- Los mensajes de DEBUG te muestran c√≥mo cambian los estados")
    print("=" * 60)

    chatbot = FlujoChatbot()

    while True:
        print(f"\nüéØ ESTADO ACTUAL: {chatbot.estado_actual}")
        mensaje_usuario = input("\nüë§ T√∫: ").strip()

        if mensaje_usuario.lower() in ['salir', 'exit', 'quit']:
            print("\nüëã ¬°Hasta luego! Gracias por usar TecnoTienda")
            break

        if not mensaje_usuario:
            print("‚ö†Ô∏è Por favor escribe algo...")
            continue

        respuesta = chatbot.procesar_mensaje(mensaje_usuario)
        print(f"\nü§ñ Bot: {respuesta}")

# Ejecutar el chatbot
ejecutar_chatbot()
````

````python
class ValidadorFlujo:
    def __init__(self):
        self.intentos_usuario = {}
        self.max_intentos = 3

    def validar_entrada(self, mensaje, tipo_validacion, usuario_id="default"):
        """Valida entrada del usuario seg√∫n el tipo requerido"""

        if usuario_id not in self.intentos_usuario:
            self.intentos_usuario[usuario_id] = 0

        self.intentos_usuario[usuario_id] += 1

        if tipo_validacion == 'email':
            if re.match(r'^[^@]+@[^@]+\.[^@]+$', mensaje):
                self.intentos_usuario[usuario_id] = 0  # Reset
                return True, mensaje

        elif tipo_validacion == 'numero_pedido':
            if re.match(r'^#?\d{6,}$', mensaje):
                self.intentos_usuario[usuario_id] = 0
                return True, mensaje.replace('#', '')

        elif tipo_validacion == 'opcion_menu':
            if mensaje in ['1', '2', '3', '4']:
                self.intentos_usuario[usuario_id] = 0
                return True, mensaje

        # Si llegamos ac√°, la validaci√≥n fall√≥
        if self.intentos_usuario[usuario_id] >= self.max_intentos:
            self.intentos_usuario[usuario_id] = 0
            return False, "Demasiados intentos fallidos. Volviendo al men√∫ principal."

        return False, f"Formato incorrecto. Intentos restantes: {self.max_intentos - self.intentos_usuario[usuario_id]}"
````
